## 1-2. df_pdf_files.pkl

### 1-2-1. 데이터 불러오기 및 확인

```python
import pandas as pd

df_pdf_files = pd.read_pickle('/content/drive/MyDrive/snowball/data/After_raw/df_pdf_files.pkl')
print(df_pdf_files.info())
print(df_pdf_files.isnull().sum())
df_pdf_files.head()
```

### 1-2-2. 컬럼 제거

- '주식수(계약수)' 컬럼과 '시가총액 구성비중' 컬럼은 추후 사용을 하지 않기에 제거

```python
df_pdf_files = df_pdf_files.drop(columns=['주식수(계약수)', '시가총액 구성비중'])
df_pdf_files.head()
```

- 시가총액 또는 평가금액 둘 중 하나에 존재하는 null 값들을 채울 때 시가총액 -> 평가금액 순으로 우선순위를 두어서 금액을 판단하기로 함

```python
# df_pdf_files에서 '시가총액'이 null인 행을 찾아서 '평가금액' 값으로 채우기
df_pdf_files['시가총액'].fillna(df_pdf_files['평가금액'], inplace=True)
print(df_pdf_files.isnull().sum())
```

```python
len(df_pdf_files.loc[(df_pdf_files['평가금액'].isnull())&(df_pdf_files['시가총액'].isnull()),:])
```

- '시가총액'의 null 값은 '평가금액'으로 채움
- 남아있는 '시가총액'의 null 값은 572275개, '평가금액'과 '시가총액'이 둘 다 null 값인 것은 572275개로 동일함을 확인

```python
# 사용안하는 '평가금액' 컬럼 제거
df_pdf_files = df_pdf_files.drop(columns=['평가금액'])
df_pdf_files.head()
```

### 1-2-3. 데이터 필터링

- df_etf_base_info의 '단축코드' 컬럼에 있는 값들로 df_pdf_files의 'etf_code' 컬럼의 값들을 필터링

```python
# df_etf_base_info의 '단축코드' 컬럼에 있는 값들을 리스트
etf_code_list = list(df_etf_base_info['단축코드'].unique())
etf_code_list[:5]
```

```python
df_pdf_files = df_pdf_files.loc[df_pdf_files['etf_code'].isin(etf_code_list),:]
print(df_pdf_files.info())
print(df_pdf_files.isnull().sum())
df_pdf_files.head()
```

- etf_code_list 의 'etf_code' 값들로 필터링을 했을 때, df_pdf_files 내의 null 값들이 제거됨을 확인

### 1-2-4. 데이터 값 변환

- '종목코드' 컬럼의 값들 중 사용하는 '종목코드'들은 6자릿수 임으로 '종목코드'가 6자리가 아닌 '종목코드'들을 확인 진행

```python
# '종목코드'가 6자리수가 아닌 종목코드들을 확인
etf_stock_code_list = list(df_pdf_files['종목코드'].unique())
cash_code_list = list()
for etf_stock_code in etf_stock_code_list:
  if len(etf_stock_code)!=6:
    cash_code_list.append(etf_stock_code)
    print(etf_stock_code, df_pdf_files.loc[df_pdf_files['종목코드']==etf_stock_code,'구성종목명'].iloc[0])
```

```python
cash_code_list[:5]
```

- 'https://m.blog.naver.com/PostView.naver?blogId=ej5824486&logNo=222544889198&categoryNo=211&proxyReferer=' 사이트를 통해,  'CASH00000001 설정현금액' 은 펀드의 설정 및 환매를 위한 현금 보유량임으로, 실질적인 주식투자 운용과는 별개로 운용사에서 펀드 관리를 위해 설정해 놓은 부분 임으로 CASH00000001 는 제거
- 'KRD010010001 원화예금' 는 원화예금을 나타내지만 다른 6자리가 아닌 '종목코드'들은 펀드등 으로 추후 가치평가를 하기 어려움, 그냥 '현금'으로 생각
-  'KRD010010001 원화예금' 의 '종목코드'는 '000000'으로, 다른 6자리가 아닌 '종목코드'들은 '999999'로 대체

```python
# 종목코드 ''CASH00000001'를 제거
df_pdf_files = df_pdf_files.loc[df_pdf_files['종목코드']!='CASH00000001',:]
df_pdf_files
```

```python
# 종목코드 'KRD010010001' 을 확인
df_pdf_files.loc[df_pdf_files['종목코드']=='KRD010010001',:]
```

```python
# 종목코드 'KRD010010001' 을 변환
df_pdf_files.loc[df_pdf_files['종목코드']=='KRD010010001','종목코드'] = '000000'
df_pdf_files.loc[df_pdf_files['종목코드']=='000000',:]
```

```python
# 종목코드가 6자리가 아닌 데이터를 확인
df_pdf_files.loc[df_pdf_files['종목코드'].isin(cash_code_list) ,:]
```

```python
# 종목코드가 6자리가 아닌 데이터의 종목코드 수정 및 확인
df_pdf_files.loc[df_pdf_files['종목코드'].isin(cash_code_list) ,'종목코드'] = '999999'
df_pdf_files.loc[df_pdf_files['종목코드']=='999999',:] # 값들이 잘 변경됨을 확인
```

```python
df_pdf_files = df_pdf_files.reset_index(drop=True)
print(df_pdf_files.info())
print(df_pdf_files.isnull().sum())
df_pdf_files.head()
```

```python
df_pdf_files.to_pickle('/content/drive/MyDrive/snowball/data/After_data_select/df_pdf_files.pkl')
```
