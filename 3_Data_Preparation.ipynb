## 1-2. df_pdf_files.pkl

### 1-2-1. 데이터 불러오기 및 확인

```python
import pandas as pd

df_pdf_files = pd.read_pickle('/content/drive/MyDrive/snowball/data/After_raw/df_pdf_files.pkl')
print(df_pdf_files.info())
print(df_pdf_files.isnull().sum())
df_pdf_files.head()
```

### 1-2-2. 컬럼 제거

- '주식수(계약수)' 컬럼과 '시가총액 구성비중' 컬럼은 추후 사용을 하지 않기에 제거

```python
df_pdf_files = df_pdf_files.drop(columns=['주식수(계약수)', '시가총액 구성비중'])
df_pdf_files.head()
```

- 시가총액 또는 평가금액 둘 중 하나에 존재하는 null 값들을 채울 때 시가총액 -> 평가금액 순으로 우선순위를 두어서 금액을 판단하기로 함

```python
# df_pdf_files에서 '시가총액'이 null인 행을 찾아서 '평가금액' 값으로 채우기
df_pdf_files['시가총액'].fillna(df_pdf_files['평가금액'], inplace=True)
print(df_pdf_files.isnull().sum())
```

```python
len(df_pdf_files.loc[(df_pdf_files['평가금액'].isnull())&(df_pdf_files['시가총액'].isnull()),:])
```

- '시가총액'의 null 값은 '평가금액'으로 채움
- 남아있는 '시가총액'의 null 값은 572275개, '평가금액'과 '시가총액'이 둘 다 null 값인 것은 572275개로 동일함을 확인

```python
# 사용안하는 '평가금액' 컬럼 제거
df_pdf_files = df_pdf_files.drop(columns=['평가금액'])
df_pdf_files.head()
```

### 1-2-3. 데이터 필터링

- df_etf_base_info의 '단축코드' 컬럼에 있는 값들로 df_pdf_files의 'etf_code' 컬럼의 값들을 필터링

```python
# df_etf_base_info의 '단축코드' 컬럼에 있는 값들을 리스트
etf_code_list = list(df_etf_base_info['단축코드'].unique())
etf_code_list[:5]
```

```python
df_pdf_files = df_pdf_files.loc[df_pdf_files['etf_code'].isin(etf_code_list),:]
print(df_pdf_files.info())
print(df_pdf_files.isnull().sum())
df_pdf_files.head()
```

- etf_code_list 의 'etf_code' 값들로 필터링을 했을 때, df_pdf_files 내의 null 값들이 제거됨을 확인

### 1-2-4. 데이터 값 변환

- '종목코드' 컬럼의 값들 중 사용하는 '종목코드'들은 6자릿수 임으로 '종목코드'가 6자리가 아닌 '종목코드'들을 확인 진행

```python
# '종목코드'가 6자리수가 아닌 종목코드들을 확인
etf_stock_code_list = list(df_pdf_files['종목코드'].unique())
cash_code_list = list()
for etf_stock_code in etf_stock_code_list:
  if len(etf_stock_code)!=6:
    cash_code_list.append(etf_stock_code)
    print(etf_stock_code, df_pdf_files.loc[df_pdf_files['종목코드']==etf_stock_code,'구성종목명'].iloc[0])
```

```python
cash_code_list[:5]
```

- 'https://m.blog.naver.com/PostView.naver?blogId=ej5824486&logNo=222544889198&categoryNo=211&proxyReferer=' 사이트를 통해,  'CASH00000001 설정현금액' 은 펀드의 설정 및 환매를 위한 현금 보유량임으로, 실질적인 주식투자 운용과는 별개로 운용사에서 펀드 관리를 위해 설정해 놓은 부분 임으로 CASH00000001 는 제거
- 'KRD010010001 원화예금' 는 원화예금을 나타내지만 다른 6자리가 아닌 '종목코드'들은 펀드등 으로 추후 가치평가를 하기 어려움, 그냥 '현금'으로 생각
-  'KRD010010001 원화예금' 의 '종목코드'는 '000000'으로, 다른 6자리가 아닌 '종목코드'들은 '999999'로 대체

```python
# 종목코드 ''CASH00000001'를 제거
df_pdf_files = df_pdf_files.loc[df_pdf_files['종목코드']!='CASH00000001',:]
df_pdf_files
```

```python
# 종목코드 'KRD010010001' 을 확인
df_pdf_files.loc[df_pdf_files['종목코드']=='KRD010010001',:]
```

```python
# 종목코드 'KRD010010001' 을 변환
df_pdf_files.loc[df_pdf_files['종목코드']=='KRD010010001','종목코드'] = '000000'
df_pdf_files.loc[df_pdf_files['종목코드']=='000000',:]
```

```python
# 종목코드가 6자리가 아닌 데이터를 확인
df_pdf_files.loc[df_pdf_files['종목코드'].isin(cash_code_list) ,:]
```

```python
# 종목코드가 6자리가 아닌 데이터의 종목코드 수정 및 확인
df_pdf_files.loc[df_pdf_files['종목코드'].isin(cash_code_list) ,'종목코드'] = '999999'
df_pdf_files.loc[df_pdf_files['종목코드']=='999999',:] # 값들이 잘 변경됨을 확인
```

```python
df_pdf_files = df_pdf_files.reset_index(drop=True)
print(df_pdf_files.info())
print(df_pdf_files.isnull().sum())
df_pdf_files.head()
```

```python
import numpy as np
from tqdm import tqdm

def get_data(index_info):


  # 업데이트 할 값들 초기화
  add_data = dict() # 추가할 데이터를 딕셔너리 형태로 생성
  add_data['종목코드'] = np.nan
  add_data['연도표시'] = np.nan
  add_data['분기표시'] = np.nan
  add_data['회사명'] = np.nan
  add_data['결산기준일'] = np.nan
  add_data['당기누적매출액'] = np.nan
  add_data['당기누적영업이익'] = np.nan
  add_data['당기누적순이익'] = np.nan

  # 값을 추출할 재무제표 생성
  df_Dart_PL_temp = df_Dart_PL_origin.loc[index_info,:].copy()

  add_data['종목코드'] = index_info[0]
  add_data['결산기준일'] = index_info[1]
  add_data['연도표시'] = list(df_Dart_PL_temp['연도표시'].unique())[0]
  add_data['분기표시'] = list(df_Dart_PL_temp['분기표시'].unique())[0]
  add_data['회사명'] = list(df_Dart_PL_temp['회사명'].unique())[0]

  # for문으로 한 row씩 돌면서 '당기누적매출액' 확인
  for index,row in df_Dart_PL_temp.iterrows():
    if row['항목코드'] in ['ifrs_Revenue','ifrs-full_Revenue']:
      add_data['당기누적매출액'] = row['분기누적금액']
      break

  # for문으로 한 row씩 돌면서 '당기누적영업이익' 확인
  for index,row in df_Dart_PL_temp.iterrows():
    if row['항목코드'] in ['dart_OperatingIncomeLoss']:
      add_data['당기누적영업이익'] = row['분기누적금액']
      break

  # for문으로 한 row씩 돌면서 '당기누적순이익' 확인
  for index,row in df_Dart_PL_temp.iterrows():
    if row['항목코드'] in ['ifrs_ProfitLoss', 'ifrs-full_ProfitLoss']:
      add_data['당기누적순이익'] = row['분기누적금액']
      break

  return add_data


# 최종적으로 만들 데이터프레임 생성(형태 생성)
df_PL = pd.DataFrame(columns = ['종목코드', '연도표시','분기표시', '회사명', '결산기준일','당기누적매출액','당기누적영업이익','당기누적순이익'])

for df_Dart_PL_origin_index in tqdm(df_Dart_PL_origin_index_list):
  # 최종적으로 조사한 데이터를 한 row 씩 추가
  df_PL = pd.concat([df_PL, pd.DataFrame([get_data(df_Dart_PL_origin_index)])], ignore_index=True)
```

#### 당기금액 계산

```python
# 두번째 row부터 for문을 돌면서 확인
for index,row in df_PL_final.loc[1:,:].iterrows():
  previous_row_stock_code = df_PL_final.loc[index-1,'종목코드']
  previous_row_quarter = df_PL_final.loc[index-1,'분기표시']

  current_row_stock_code = row['종목코드']

  # 새로운 '종목코드' 정보가 나오면 다음 코드들을 실행하지 X
  if previous_row_stock_code != current_row_stock_code:
    continue


  if previous_row_quarter == 4: # 이전 row의 분기표시가 4 이면 새롭게 값들을 계산 시작
    df_PL_final.loc[index,'당기매출액'] = row['당기누적매출액']
    df_PL_final.loc[index,'당기영업이익'] = row['당기누적영업이익']
    df_PL_final.loc[index,'당기순이익'] = row['당기누적순이익']
  else: # 나머진 경우는 현재 누적값에서 이전 row의 누적값을 빼서 당기금액을 계산
    previous_row_revenue = df_PL_final.loc[index-1,'당기누적매출액']
    previous_row_operating = df_PL_final.loc[index-1,'당기누적영업이익']
    previous_row_profit = df_PL_final.loc[index-1,'당기누적순이익']

    df_PL_final.loc[index,'당기매출액'] = row['당기누적매출액'] - previous_row_revenue
    df_PL_final.loc[index,'당기영업이익'] = row['당기누적영업이익'] - previous_row_operating
    df_PL_final.loc[index,'당기순이익'] = row['당기누적순이익'] - previous_row_profit
```
